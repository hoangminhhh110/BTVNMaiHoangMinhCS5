```python
def nguyento(n):
    if n == 1: 
        return False
    for i in range (2, n):
       if n % i == 0: 
           return False 
    return True
n = int(input("Nhập số nguyên dương n: "))
tổng  = 0
for  i in range (1, n+1):
    if n % i == 0 and nguyento(i) == True: 
        tổng = max(tổng, i)
print(tổng)
```
